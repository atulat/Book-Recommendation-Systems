```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
%%time
similarity = cosine_similarity(sparse_matrix)
```

```python
from collections import Counter
```

```python
def recommend_books_count_new(user_id, top_k=5):
    if user_id not in pivot.index:
        return "The entered User-ID is not in the Database"
    
    index = np.where(pivot.index==user_id)[0][0]
    similar_items = sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse=True)[1:top_k+1]
    
    similar_users = [pivot.index[i[0]] for i in similar_items]
    
    books = []
    for similar_user in similar_users:
        books += user_book_map[similar_user]
    
    books = Counter(books)
    books = books.most_common(10)
    
    recommended_books = [book[0] for book in books if pivot.loc[user_id, book[0]] == 0]
    
    if len(recommended_books) < 5:
        for book in pivot.columns:
            if book not in recommended_books and pivot.loc[user_id, book] == 0:
                recommended_books.append(book)
                if len(recommended_books) >= 5:
                    break
    
    return recommended_books[:3]
```

```python
def excel(user_id, top_k=10):
    index = np.where(pivot.index==user_id)[0][0]
    similar_items = sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse=True)[1:top_k+1]
    
    similar_users = [pivot.index[i[0]] for i in similar_items]
    
    books = []
    for similar_user in similar_users:
        books += user_book_map[similar_user]
    
    books = Counter(books)
    books = books.most_common(10)
    
    recommended_books = [book[0] for book in books]
    
    if len(recommended_books) < 10:
        difference = 10 - len(recommended_books)
        for book in pivot.columns:
            if book not in recommended_books and difference > 0:
                recommended_books.append(book)
                difference -= 1
    
    result = []
    for book in recommended_books:
        rating = df[(df['User-ID'] == user_id) & (df['bookTitle'] == book)]['bookRating'].values
        if len(rating) > 0:
            result.append((book, rating[0]))
        else:
            result.append((book, 0))
    
    return result
```

```python
def recommend_books_dep(user_input, top_k=5):
    if user_input not in pivot_1.index:
        return "The entered User-ID is not in the Database"
    
    index = np.where(pivot_1.index==user_input)[0][0]
    similar_items = sorted(list(enumerate(similarity_1[index])),key=lambda x:x[1],reverse=True)[1:top_k+1]
    
    similar_users = [pivot_1.index[i[0]] for i in similar_items]
    
    books = []
    for similar_user in similar_users:
        books += user_book_map[similar_user]
    
    books = Counter(books)
    books = books.most_common(10)
    
    recommended_books = [book[0] for book in books if pivot_1.loc[user_input, book[0]] == 0]
    
    output = []
    for i in recommended_books:
        item = []
        temp_df = image[image['Book-Title'] == i]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))
        output.append(item)
        
    return output[:3]
```
