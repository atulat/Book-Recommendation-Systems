## Using Popularity Based Approach to display top 50 books from the Dataset

```python
popular = df.groupby('bookTitle').count()['bookRating'].reset_index()
popular.rename(columns={'bookTitle':'Title','bookRating':'Votes'},inplace=True)
popular
```

```python
rating = df.groupby('bookTitle').mean()['bookRating'].reset_index()
rating.rename(columns={'bookTitle':'Title','bookRating':'Rating'},inplace=True)
rating
```

```python
popular = popular.merge(rating,on='Title')
popular
```

```python
popular = popular[popular['Votes']>=200].sort_values('Rating',ascending=False).head(50)
popular.rename(columns={'Title':'Book-Title'},inplace=True)
popular
```

```python
popular = popular.merge(image,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','Votes','Rating']]
```

```python
popular.rename(columns={'Book-Title':'Title','Book-Author':'Author','Year-Of-Publication':'Release_year','Image-URL-M':'Image'},inplace=True)
```

```python
popular['Rating'] = popular['Rating'].round(2)
```

## Item-Item Based Collaborative Filtering

```python
import pickle
pickle.dump(popular,open('popular.pkl','wb'))
```

```python
x = df.groupby('User-ID').count()['bookRating'] > 200
filtered_users = x[x].index
```

```python
filtered = df[df['User-ID'].isin(filtered_users)]
filtered
```

```python
y = df.groupby('bookTitle').count()['bookRating'] > 100
filtered_books = y[y].index
```

```python
filtered = df[df['bookTitle'].isin(filtered_books)]
filtered
```

```python
filtered.rename(columns={'User-ID':'user_id'},inplace=True)
filtered
```

```python
filtered.duplicated().sum()
```

```python
pickle.dump(filtered,open('filtered_item.pkl','wb'))
```

```python
pivot = filtered.pivot_table(index='bookTitle',columns='user_id',values='bookRating')
pivot.fillna(0,inplace=True)
```

```python
pivot
```

```python
pickle.dump(pivot,open('pivot_item.pkl','wb'))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
%%time
similarity = cosine_similarity(pivot)
```

```python
pickle.dump(similarity,open('similarity_item.pkl','wb'))
```

```python
similarity.shape
```

```python
def recommend(book_name):
    index = np.where(pivot.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pivot.index[i[0]])
```

```python
recommend('Harry Potter and the Chamber of Secrets (Book 2)')
```

```python
recommend('The Fellowship of the Ring (The Lord of the Rings, Part 1)')
```

```python
def recommend_dep(book_name):
    index = np.where(pivot.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    output = []
    for i in similar_items:
        item = []
        temp_df = image[image['Book-Title'] == pivot.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))
        output.append(item)
        
    return output
```

```python
recommend_dep('1984')
```
